## Upload and Convert

### Provide Animated Drawing URL
Copy and paste the url below

```python
url = "https://sketch.metademolab.com/share/f52a371b84ed4a95a5201400e2339796/hip_hop_dancing"
```

### Download the mp4 file

```python
# Download the mp4 file

import re
import requests

def download(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print("Download complete.")
    else:
        print("Failed to download the file.")

match = re.search(r"share/(.*?)/(.*)", url)
if match:
    id = match.group(1)
    dancing_mode = match.group(2)
    print(f"{id} {dancing_mode}")
    mp4_url = f"https://production-sketch-video.metademolab.com/{id}/{dancing_mode}.mp4"

    download(mp4_url, "input.mp4")
```

### Remove background

```python
!backgroundremover -i input.mp4 -tg -o output.gif
```
